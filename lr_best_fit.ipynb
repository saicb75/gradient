
### Steps for computing gradient descent

1. step 1 : Random initialization of w0 & w1
2. step 2 : Compute ypred
3. step 3 : Compute cost
4. step 4 : check if cost is near zero then exit 
5. step 5 : Re-initialize w0 & w1
6. step 6 : Go to step2

```python
# importing the library
import matplotlib.pyplot as plt
```

```python
# function to compute the cost 

def cost_func(ya,yp):
    
    ys =[]
    tot = 0
    
    for i in range(0,len(ya)):
        ys.append((ya[i]-yp[i])**2)
        tot+=ys[i]
    
    cost = tot/len(ys)    
    return cost  
```

```python
# function to reinitialize the weights

def reinit_wt(x, w0, w1, ya, yp, alpha):
    
    # formula used for computing w0, w1
    # w0 = w0 - 1/m . α . ∑(hθ(xi)−yi
    # w1 = w1 − 1/m . α . ∑((hθ(xi)−yi)xi)
    
    temp_w0 = 0
    temp_w1 = 0
    
    for i in range(0,len(ya),1):
        temp_w0 += (ya[i]-yp[i])
        temp_w1 += (ya[i]-yp[i])*x[i]
        
    mean_residual_w0 = temp_w0 / len(ya)
    mean_residual_w1 = temp_w1 / len(ya)
    w0 = w0 - (alpha*mean_residual_w0)
    w1 = w1 - (alpha*mean_residual_w1)
    
    return w0, w1
```

```python
# given vales of x and y

x = [1, 2, 3]
ya = [1, 4, 3]
```

```python
# step 1 : Random initialization of w0 & w1

w0 = 0
w1 = 1
alpha = -0.01
```

```python
cost_curr = 0
cost_prev = 0

for iter_new in range(0, 10000):
    
    # step 2 : Compute ypred
    # ypred = w0 + w1*x

    ypred = []
    for i in range(0, len(x)):
        ypred.append(w0 + (w1 * x[i]))

    print("y_pred", ypred)
   
    # step 3 : Compute cost
    # cost = mean(residual ** 2), using RMSE as cost function

    cost_curr = cost_func(ya, ypred)
    
    print("Current cost is : ", cost_curr)
    print("Number of iteration is : ", iter_new) 
    print("The cost diff is {}".format(abs(cost_curr - cost_prev)))
       
    # step 4 : check if cost is near zero then exit 
    
    if (abs(cost_curr - cost_prev)) < 0.0000001:
        break

    cost_prev = cost_curr

    # step 5 : Re-initialize w0 & w1

    (w0, w1) = reinit_wt(x, w0, w1, ya, ypred, alpha)

    print("new weights are {} {}" .format(w0, w1))
    print(" ")

eqn = "y = {}*x + {}".format(round(w1, 2), round(w0, 3))
print("best fit line: {}".format(eqn))
print("cost = {}".format(cost_curr))
print("ypred = {}".format(ypred))
print("y = {}".format(ya))
```

```python
# Plotting the target, predicted values and the regression line

plt.scatter(x, ya, label='actual' )
plt.scatter(x, ypred, c='r', label='predicted' )
plt.plot(x, ypred, c='r', label="regression line")

plt.title(label = "x vs. y, ypred")
plt.xlabel(xlabel = "x" )
plt.ylabel(ylabel = "y and ypred" )
plt.text(x=1.5, y=2.0, s=f'{eqn}')
plt.legend()
```
